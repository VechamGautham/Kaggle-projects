## Correlation Graph

```python
correlations = train_df.corr(numeric_only=True)["Calories"].drop("Calories").sort_values(ascending=True)
plt.xlabel("Correlation with calories")
plt.ylabel("Features")
plt.title("Correlation with Calories")
plt.barh(correlations.index, correlations.values)
plt.show()
```

# Correlation graph with respect to Calories without feature engineering 

```python
def scatter_plot_corr(df):
    cols = ["Calories"] + [col for col in df.columns if col != "Calories"]
    sns.pairplot(df[cols],y_vars=["Calories"],x_vars=cols[2:], kind="scatter", diag_kind="hist")   
    plt.show()
```

```python
scatter_plot_corr(train_df)
```

# Correlation value with respect to calories without feature engineering 

```python
def corr_val(df):
    return df.corr(numeric_only=True)["Calories"].drop("Calories").sort_values(ascending=False)
```

```python
corr_val(train_df)
```

# FEATURE ENGINEERING

```python
temp_df = train_df.copy()
# BMI = weight (kg) / [height (m)]
temp_df["BMI"] = temp_df["Weight"] / ((temp_df["Height"] / 100) ** 2)   
temp_df["BMI"] = temp_df["BMI"].round(2)
temp_df["Duration_weight"] = temp_df["Duration"] * temp_df["Weight"]
temp_df["intensity"] = temp_df["Heart_Rate"] / temp_df["Calories"]
temp_df["Weight_per_age"] = temp_df["Weight"] / temp_df["Age"]
temp_df["Weight_per_Height"] = temp_df["Weight"] / temp_df["Height"]
temp_df["intensity_score"] = (temp_df["Heart_Rate"]*temp_df["Body_Temp"])/temp_df["Duration"]
temp_df["Calories_per_minute"] = temp_df["Calories"] / temp_df["Duration"]
temp_df["heart_rate_per_kg"] = temp_df["Heart_Rate"] / temp_df["Weight"]
temp_df["heart_reate_per_age"] = temp_df["Heart_Rate"] / temp_df["Age"]
temp_df["BSA"] = 0.007184 * (temp_df["Weight"] ** 0.425) * (temp_df["Height"] ** 0.725)
temp_df["BSA"] = temp_df["BSA"].round(2)
temp_df["temp_heart_product"] = temp_df["Body_Temp"] * temp_df["Heart_Rate"]
temp_df["duration_per_age"] = temp_df["Duration"] / temp_df["Age"]


temp_df["Sex_encoded"] = temp_df["Sex"].map({"male": 0, "female": 1})
```

```python
scatter_plot_corr(temp_df)
```

```python
corr_val(temp_df)
```
