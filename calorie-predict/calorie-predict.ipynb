# Correlation graph with respect to Calories without feature engineering 

```python
def scatter_plot_corr(df):
    cols = ["Calories"] + [col for col in df.columns if col != "Calories"]
    sns.pairplot(df[cols],y_vars=["Calories"],x_vars=cols[2:], kind="scatter", diag_kind="hist")   
    plt.show()
```

```python
scatter_plot_corr(train_df)
```

# Correlation value with respect to calories without feature engineering 

```python
def corr_val(df):
    return df.corr(numeric_only=True)["Calories"].drop("Calories").sort_values(ascending=False)
```

# FEATURE ENGINEERING

```python
temp_df = train_df.copy()
temp_test_df = test_df.copy()
# BMI = weight (kg) / [height (m)]
temp_df["BMI"] = temp_df["Weight"] / ((temp_df["Height"] / 100) ** 2)   
temp_df["BMI"] = temp_df["BMI"].round(2)
temp_test_df["BMI"] = test_df["Weight"] / ((test_df["Height"] / 100) ** 2)
temp_test_df["BMI"] = temp_df["BMI"].round(2)

# duration * weight
temp_df["Duration_weight"] = temp_df["Duration"] * temp_df["Weight"]
temp_test_df["Duration_weight"] = temp_test_df["Duration"] * temp_test_df["Weight"]


# weight per age
temp_df["Weight_per_age"] = temp_df["Weight"] / temp_df["Age"]
temp_test_df["Weight_per_age"] = temp_test_df["Weight"] / temp_test_df["Age"]

# weight per height
temp_df["Weight_per_Height"] = temp_df["Weight"] / temp_df["Height"]
temp_test_df["Weight_per_Height"] = temp_test_df["Weight"] / temp_test_df["Height"]

# intensity score = (heart_rate * body_temp) / duration
temp_df["intensity_score"] = (temp_df["Heart_Rate"]*temp_df["Body_Temp"])/temp_df["Duration"]
temp_test_df["intensity_score"] = (temp_test_df["Heart_Rate"]*temp_test_df["Body_Temp"])/temp_test_df["Duration"]



# heart rate per kg
temp_df["heart_rate_per_kg"] = temp_df["Heart_Rate"] / temp_df["Weight"]
temp_test_df["heart_rate_per_kg"] = temp_test_df["Heart_Rate"] / temp_test_df["Weight"]

# heart rate per age
temp_df["heart_reate_per_age"] = temp_df["Heart_Rate"] / temp_df["Age"]
temp_test_df["heart_reate_per_age"] = temp_test_df["Heart_Rate"] / temp_test_df["Age"]

# body surface area (BSA) = 0.007184 * (weight^0.425) * (height^0.725)
temp_df["BSA"] = 0.007184 * (temp_df["Weight"] ** 0.425) * (temp_df["Height"] ** 0.725)
temp_df["BSA"] = temp_df["BSA"].round(2)
temp_test_df["BSA"] = 0.007184 * (temp_test_df["Weight"] ** 0.425) * (temp_test_df["Height"] ** 0.725)
temp_test_df["BSA"] = temp_test_df["BSA"].round(2)

# temperature * heart rate
temp_df["temp_heart_product"] = temp_df["Body_Temp"] * temp_df["Heart_Rate"]
temp_test_df["temp_heart_product"] = temp_test_df["Body_Temp"] * temp_test_df["Heart_Rate"]

# duration per age
temp_df["duration_per_age"] = temp_df["Duration"] / temp_df["Age"]
temp_test_df["duration_per_age"] = temp_test_df["Duration"] / temp_test_df["Age"]

# sex encoding
temp_df["Sex_encoded"] = temp_df["Sex"].map({"male": 0, "female": 1})
temp_test_df["Sex_encoded"] = temp_test_df["Sex"].map({"male": 0, "female": 1})
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_squared_log_error, r2_score

x = temp_df.drop(columns=["id","Calories","Sex"])
y = temp_df["Calories"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
xgb = XGBRegressor(n_estimators=1000, learning_rate=0.01, max_depth=3, random_state=42)
xgb.fit(x_train, y_train)
y_pred = xgb.predict(x_test)
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("RMSLE:", np.sqrt(mean_squared_log_error(y_test, y_pred)))
print("R2:", r2_score(y_test, y_pred))  
```
